# Fixing the Resolution for High-DPI Screens

The last thing you want when creating your visualizations is a blurry image. Unfortunately, that's exactly what you get by default when you use the ggplot2 library within a Jupyter notebook. One option for fixing this is to go with PNG output and set a resolution and pair of image dimensions that work well with your high-DPI screen. A typical PPI for an Apple "Retina" display is around 220ppi (give or take), so we'll set our image resolution to that. 

Then, we need to choose a pair of image dimensions. Choosing something too small means we end up with the components of the graph (e.g., the text, dots, lines, etc.) being a bit too big. 9x7 inches seems to be a good fit for a PPI of 220. The following code can be added to the top of your Jupyter Notebook to set the image output defaults for the entire notebook.

```r
options(jupyter.plot_mimetypes = "image/png",
        repr.plot.width        = 9,
        repr.plot.height       = 7,
        repr.plot.res          = 220)
```

This works well and creates nice clean graphs graphs on high-DPI screens, but another option, that I quite like is to simply go with the SVG format instead of PNG. Unlike the first option, going with SVG over PNG allows us to choose any size for our image output without worrying about the size of the visualization components. So, we can go as small or as big as we want with our images. Personally, I find a 7x5 inch image to be just about the perfect size, so we'll setup our defaults to output a 7x5 inch SVG image.

````r
# Using svg+xml, instead of png, produces a much clearer/crisper 
# image at any size on high-DPI screens such as retina displays.
options(jupyter.plot_mimetypes    = "image/svg+xml",
        repr.plot.width           = 7, 
        repr.plot.height          = 5)
````
